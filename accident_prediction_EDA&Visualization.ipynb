```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\vishn\Downloads\accident_prediction_india.csv")
```

1. Accidents by Weather Condition (Pie Chart)

```python
weather_counts = df['Weather condition'].value_counts()
plt.figure(figsize=(7,7))
weather_counts.plot.pie(autopct='%1.1f%%', startangle=140, fontsize=13)
plt.title('Accidents by Weather Condition')
plt.ylabel('')
plt.show()
```

2. Accidents by Road Condition (Pie Chart)

```python
road_counts = df['Road Condition'].value_counts()
plt.figure(figsize=(7,7))
road_counts.plot.pie(autopct='%1.1f%%', startangle=140, fontsize=13)
plt.title('Accidents by Road Condition')
plt.ylabel('')
plt.show()
```

3. Accidents by Time of Day (Bar Chart)

```python
time_counts = df['Time of Day'].value_counts()
plt.figure(figsize=(8,5))
time_counts.plot(kind='bar', color='skyblue')
plt.title('Accidents by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=30)
plt.show()
```

4. Top 10 Cities with Most Accidents (Bar Chart)

```python
city_counts = df['City Name'].value_counts().head(10)
plt.figure(figsize=(10,6))
city_counts.plot(kind='bar', color='orange')
plt.title('Top 10 Cities by Number of Accidents')
plt.xlabel('City')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=25)
plt.show()
```

5. Accident Severity Distribution (Bar Chart)

```python
severity_counts = df['Accident Severity'].value_counts()
plt.figure(figsize=(7,5))
severity_counts.plot(kind='bar', color='red')
plt.title('Accident Severity Distribution')
plt.xlabel('Severity')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=15)
plt.show()
```
